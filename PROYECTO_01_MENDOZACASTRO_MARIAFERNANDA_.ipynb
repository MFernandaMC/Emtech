```python
#Importacion de librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as skl
```

```python
#Se guardo el archivo csv y se guardara su ruta en la variable llamada dataframe
dataframe=pd.read_csv('/content/country-income.csv')

#head sirve para dicir cuantas filas queremos mostrar
dataframe.head(10)
```

```python
#.info sirve para crear un resumen del dataframe 
dataframe.info()
#no se encontraron valores nulos 
#cuenta con 10 filas y 4 columnas 
```

```python
#.fillna sustituye los valores perdidos por la media de los datos
dataframe.fillna(dataframe.mean())
#no se sustituyo ningun dato porque no hay datos perdidos 
```

```python
#.mean sirve para calcular la media 
#.max sirve para calcular el maximo
#.min sirve para calcular el minimo
#.std() sirve para calcular la desviacion

#Calculo de la media, maxima, minima y desviacion de la columna Age
print ('Datos de la columna Age')
print ('La media es', dataframe['Age'].mean(), ', La maxima es', dataframe['Age'].max(), 
       ', El minimo es', dataframe['Age'].min(), 'y La desviacion es de', dataframe['Age'].std(),  )

print ('')

#Calculo de la media, maxima, minima y desviacion de la columna Income
print ('Datos de la columna Income')
print ('La media es', dataframe['Income'].mean(), ', La maxima es', dataframe['Income'].max(), 
       ', El minimo es', dataframe['Income'].min(), 'y La desviacion es de', dataframe['Income'].std(),  )
```
